```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df=pd.read_csv("/content/data1.csv")
print(df)
```

```python
data=list(zip(df['Duration'],df['Pulse']))
kmeans = KMeans(n_clusters=5, max_iter=50)
kmeans.fit(data)
plt.scatter(df['Duration'],df['Pulse'], c=kmeans.labels_)
plt.show()
```
