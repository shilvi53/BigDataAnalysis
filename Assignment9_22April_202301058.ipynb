```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'smoking_history': [4.9, 4.7, 4.6, 5, 5.4, 4.6, 5, 4.4, 4.9, 5.4],
    'diabetes': [3, 3.2, 3.1, 3.6, 3.9, 3.4, 3.4, 2.9, 3.1, 3.7]
})
threshold = 3.5
data['diabetes_binary'] = (data['diabetes'] > threshold).astype(int)
X = data[['smoking_history']]
Y = data['diabetes_binary']
model = LogisticRegression(solver='liblinear')
model.fit(X, Y)
predicted_probs = model.predict_proba(X)[:, 1]
plt.scatter(X, Y)
plt.plot(X, predicted_probs, color='red')
plt.xlabel('Smoking History')
plt.ylabel('Probability of Diabetes')
plt.title('Logistic Regression Model')
plt.show()
```
